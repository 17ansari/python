```python
df["Turnover"]=df["Avg. price"]*df["Qty."]
df.head(2)
```

```python
df["Brokerage"]=np.where(df["Status"] =="COMPLETE",20, 0)
df.head(2)
```

```python
t=df["Turnover"]*0.00025
```

```python
df["STT/CTT"]=np.where((df["Status"] =="COMPLETE") & (df["Type"]=="SELL"),t.round(0),0)
df.head(2)
```

```python
df["ETC"]=0.0000325*df["Turnover"]
df["ETC"]=df["ETC"].round(2)
```

```python
seb = (df["Turnover"]*10/10000000)
```

```python
df["SEBI"]=np.where(df["Status"]=="COMPLETE",seb,0)
df.head()
```

```python
bst=df["Brokerage"]+df["SEBI"]+df["ETC"]
bst1=bst*(0.18)
```

```python
df["GST"]=np.where(df["Status"]=="COMPLETE",bst1.round(2),0)
df.head(2)
```

```python
df["Stamp charges"]=np.where((df["Type"]=="BUY")&(df["Status"]=="COMPLETE"),(df["Turnover"]*0.00003).round(2),0)
df.head(2)
```

```python
df["Total charges"]=df["Brokerage"]+df["STT/CTT"]+df["SEBI"]+df["GST"]+df["Stamp charges"]+df["ETC"]
df.head(2)
```

```python
stockwise = df.groupby(["Instrument","Type"])[["Qty.","Turnover","Brokerage","STT/CTT","ETC","SEBI","Stamp charges","GST","Total charges"]].sum()
```

```python
stockwise
```

```python
buy_turnover = df[df['Type'] == 'BUY'].groupby('Instrument')['Turnover'].sum().reset_index()
sell_turnover = df[df['Type'] == 'SELL'].groupby('Instrument')['Turnover'].sum().reset_index()

# Rename columns for clarity
buy_turnover.columns = ['Instrument', 'Buy Turnover']
sell_turnover.columns = ['Instrument', 'Sell Turnover']

# Merge the buy and sell turnover dataframes
turnover_df = pd.merge(buy_turnover, sell_turnover, on='Instrument', how='outer').fillna(0)

# Calculate Gross PnL
turnover_df['Gross PnL'] = turnover_df['Sell Turnover'] - turnover_df['Buy Turnover']

# Display the turnover dataframe with Gross PnL
total_charges = df.groupby('Instrument')['Total charges'].sum().reset_index()
turnover_df = pd.merge(turnover_df, total_charges, on='Instrument', how='outer').fillna(0)

# Calculate Net PnL
turnover_df['Net PnL'] = turnover_df['Gross PnL'] - turnover_df['Total charges']

# Calculate % Charges on Gross PnL
turnover_df['% Charges on Gross PnL'] = (turnover_df['Total charges'] / turnover_df['Gross PnL']) * 100

# Display the overall summary dataframe
print(turnover_df)
```

```python
with pd.ExcelWriter("KitePnL.xlsx") as writer:
    df.to_excel(writer, sheet_name="Individual Transaction charges", index=False)
    stockwise.to_excel(writer, sheet_name="Stockwise & Typewise summary", index=False)
    turnover_df.to_excel(writer, sheet_name="Stockwise summary", index=False)
```
